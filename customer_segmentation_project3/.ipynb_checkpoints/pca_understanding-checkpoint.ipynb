```python
########### using correlation matrix to compute eigenvectors and eigen values
cor_mat1 = np.corrcoef(X_std.T)
eig_vals, eig_vecs = np.linalg.eig(cor_mat1)

print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
eig_pairs= [(eig_vals[i],list(eig_vecs[:,int(i)])) for i in range(len(eig_vals))]
```

```python
eig_pairs.sort(key= lambda k:k[0], reverse=True)
```

```python
total_sum_eig_vals = np.sum(list(eig_vals))
```

```python
total_variance = [eig_vals[i]/total_sum_eig_vals*100 for i in range(len(sorted(eig_vals,reverse=True)))]
```
